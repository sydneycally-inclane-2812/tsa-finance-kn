```python
import yfinance as yf
to_get = ['AMD', 'NVDA', 'INTC']
data = yf.download(to_get, start='2018-01-01', end='2023-12-31')
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
import pandas as pd
import numpy as np

pf = data['Close']
returns = pf.pct_change()
weights = np.array([0.3, 0.3, 0.3])
pf["Portfolio"] = returns.dot(weights)
```

```python
returns.to_csv('returns.csv')
returns
```

```python
returns.describe()
returns.isna().sum()
```

```python
pf
```

```python
import matplotlib.pyplot as plt
# Calculate absolute returns and squared returns
pf['Absolute Returns'] = pf['Portfolio'].abs()
pf['Squared Returns'] = pf['Portfolio'] ** 2
plt.figure(figsize=(12, 8))

# Portfolio returns plot
plt.subplot(3, 1, 1)
pf['Portfolio'].plot(title='Portfolio Returns', xlabel='Date', ylabel='Returns')
plt.subplot(3, 1, 2)
pf['Absolute Returns'].plot(title='Portfolio Absolute Returns', xlabel='Date', ylabel='Returns')
plt.subplot(3, 1, 3)
pf['Squared Returns'].plot(title='Portfolio Squared Returns', xlabel='Date', ylabel='Returns')
plt.tight_layout()
plt.show()
# Descriptive statistics
print(pf[['Portfolio', 'Absolute Returns', 'Squared Returns']].describe())

# ACF plot
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(pf['Portfolio'].dropna())
plt.title('Autocorrelation Function')
plt.show()
```

```python
# Comments on the plots:
```

```python
from scipy.stats import shapiro
from statsmodels.stats.diagnostic import acorr_ljungbox

# Normality test (Shapiro-Wilk)
shapiro_test_stat, shapiro_p_value = shapiro(pf['Portfolio'].dropna())
print("Shapiro-Wilk Test:")
print("Test statistic:", shapiro_test_stat)
print("p-value:", shapiro_p_value)
if shapiro_p_value < 0.05:
    print("The series is not normally distributed according to the Shapiro-Wilk test.")
else:
    print("The series is normally distributed according to the Shapiro-Wilk test.")

# Ljung-Box test
ljung_box = acorr_ljungbox(pf['Portfolio'].dropna(), lags=20)
print("\nLjung-Box Test:")
print("Test statistics:", ljung_box['lb_stat'])
print("p-values:", ljung_box['lb_pvalue'])
if any(p < 0.05 for p in ljung_box['lb_pvalue']):
    print("The series is not independently distributed according to the Ljung-Box test.")
else:
    print("The series is independently distributed according to the Ljung-Box test.")
```

```python
import arch
from arch import arch_model

# Create returns
returns = pf['Portfolio'].copy().dropna()

# Fit EWMA model
ewma_model = arch_model(returns, vol='GARCH', p=1, o=0, q=0, dist='t')
ewma_model_fit = ewma_model.fit(disp='off')
print('EWMA Model:')
print(ewma_model_fit.summary())
```

```python
returns.to_csv('returns.csv')
```

```python
# Fit GARCH(1,1) model
garch_model = arch_model(returns, vol='GARCH', p=1, o=0, q=1, dist='t')
garch_model_fit = garch_model.fit(disp='off')
print('\nGARCH(1,1) Model:')
print(garch_model_fit.summary())
```

```python
# Fit GJR-GARCH(1,1) model
gjr_garch_model = arch_model(returns, vol='GARCH', p=1, o=1, q=1, dist='t')
gjr_garch_model_fit = gjr_garch_model.fit(disp='off')
print('\nGJR-GARCH(1,1) Model:')
print(gjr_garch_model_fit.summary())
```

```python
# Comments on the models:
```

```python
models = [ewma_model_fit, garch_model_fit, gjr_garch_model_fit]
models_names = ['EWMA', 'GARCH(1,1)', 'GJR-GARCH(1,1)']

aic_scores = [model.aic for model in models]
sbic_scores = [model.bic for model in models]

print("\nAIC Scores:", aic_scores)
print("SBIC Scores:", sbic_scores)

print("\nBest model according to AIC:", models_names[aic_scores.index(min(aic_scores))])
print("Best model according to SBIC:", models_names[sbic_scores.index(min(sbic_scores))])

print("\nAll AIC scores:")
for i in range(len(models)):
    print(models_names[i], ":", aic_scores[i])
print("\nAll SBIC scores:")
for i in range(len(models)):
    print(models_names[i], ":", sbic_scores[i])
```
